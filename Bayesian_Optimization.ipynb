```python
import sklearn 
import sklearn.preprocessing 
import sklearn.decomposition
import sklearn.utils 
import sklearn.gaussian_process
import sklearn.metrics 
import scipy
import scipy.stats 
import numpy as np 
import csv

# .values gets rid of the initial indexing in front of the numbers 
Xdata = df_Solvent_Features_All.values  
print(Xdata)
Xdatainfo = df['Reactant'].values
print(Xdatainfo) 
Ydata = df['Potential'].values
Ydata = -1 * np.abs(np.subtract(df['Potential'], np.full(Ydata.shape, 1.5500)))
Ydata = Ydata.reset_index(drop=True)
print(Ydata)

scaler = sklearn.preprocessing.StandardScaler()
Xdata = scaler.fit_transform(Xdata)

n_components = 15
pca = sklearn.decomposition.PCA(n_components=n_components)
Xdata = pca.fit_transform(Xdata)

Xdata, Xdatainfo, Ydata = sklearn.utils.shuffle(Xdata, Xdatainfo, Ydata)

n_data = len(Ydata) 
n_features = len(Xdata[0, :]) 

def gpr_model(Xtrain, Ytrain): 
    nPC = 15 # number of principal components (from PCA)
    cmean=[1.0]*nPC
    cbound=[[1e-3, 1e3]]*nPC
    kernel = C(1.0, (1e-3, 1e3)) * matk(cmean, cbound, 1.5)+ Wht(1.0, (1e-3, 1e3))
    n_restarts_optimizer = 10
    gpr = sklearn.gaussian_process.GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=n_restarts_optimizer, normalize_y=False)
    gpr = gpr.fit(Xtrain, Ytrain)
    return gpr

def gpr_prediction(Xdata, gpr): 
    prediction, sigma = gpr.predict(Xdata, return_std=True)
    return prediction, sigma 

def expected_improvement(X_data, y_best, gpr): 
    y_pred, sigma = gpr_prediction(Xdata, gpr) 
    ei = []
    expl = 0.01
    for i in range(len(y_pred)): 
        Z = (y_pred[i] - y_best - expl)/sigma[i]
        expected_i = (y_pred[i] - y_best)*scipy.stats.norm.cdf(Z) + sigma[i]*scipy.stats.norm.pdf(Z)
        ei.append(expected_i)
    return ei

n_train = 5
def find_best_mol(Xdata, Xdatainfo, Ydata, n_iterations): 
    Xdata = pd.DataFrame(Xdata) 
    Xdatainfo = pd.DataFrame(Xdatainfo)
    Ydata = pd.DataFrame(Ydata)
    
    n_train = 5
    Xtrain = Xdata.iloc[:5, :]
    Xtraininfo = Xdatainfo.iloc[:5]
    Ytrain = Ydata.iloc[:5]
    
    XtrainInit = Xtrain 
    XtraininfoInit = Xtraininfo 
    YtrainInit = Ytrain 
    
    yoptloc = np.argmax(Ytrain)
    yoptval = Ytrain.iloc[yoptloc]
    xoptval = Xtraininfo.iloc[yoptloc] 
    yoptval_int = yoptval
    xoptval_int = xoptval
    
    Xremain = Xdata.iloc[5:,:] 
    Xremaininfo = Xdatainfo.iloc[5:] 
    Yremain = Ydata.iloc[5:].reset_index(drop=True) ##
    
    mae = [] # mean average error 
    rmse = [] # root mean squared error
    Xinfoexplored = []
    Yexplored = []
    
    N_iterations = n_iterations
    
    for i in range(N_iterations): 

        gpr = gpr_model(Xtrain, Ytrain)
        ypreds, sigma = gpr_prediction(Xtrain, gpr) 
        ybestloc = np.argmax(ypreds)
        # problem is that the index number for molecule here is different from the actual index number within the train array it was placed in -> reset index every time 
        Xtrain = pd.DataFrame(Xtrain)
        Ytrain = pd.DataFrame(Ytrain)
        xbest = Xtrain.iloc[ybestloc,:] 
        ytrue = Ytrain.iloc[ybestloc]
        
        if int(ytrue) > int(yoptval): 
            yoptval = ytrue 
            xoptval = xbest 
        
        rmse_temp = sklearn.metrics.mean_squared_error(Ytrain, ypreds)
        mae_temp = sklearn.metrics.mean_absolute_error(Ytrain, ypreds)
        rmse.append(rmse_temp) 
        mae.append(mae_temp)
        
        ei = expected_improvement(Xremain, yoptval, gpr)
        eioptloc = np.argmax(ei) ## next identify the best molecule to explore
        y_next_best = Yremain.iloc[eioptloc] 
        x_next_best = Xremain.iloc[eioptloc] 
        xinf_next_best = Xremaininfo.iloc[eioptloc]
        # remember that gpr.predict takes in 2D arrays as input features to make prediction based on 
        print("The next best molecule is ", xinf_next_best, " with predicted reduction potential of ", gpr.predict([x_next_best]), " . The current best has potential of ", yoptval)
        
        Xtrain_new = Xtrain.append(Xremain.iloc[eioptloc])
        Xtraininfo_new = Xtraininfo.append(Xremaininfo.iloc[eioptloc])
        Ytrain_new = Ytrain.append(Yremain.iloc[eioptloc])
        Xremain_new = Xremain.drop(eioptloc, axis=0)
        Xremaininfo_new = Xremaininfo.drop(eioptloc, axis=0) 
        Yremain_new = Yremain.drop(eioptloc, axis=0)
        
        del Xtrain
        del Xtraininfo
        del Ytrain
        del Xremain 
        del Xremaininfo
        del Yremain 
        del gpr
        
        Xtrain = Xtrain_new 
        Xtraininfo = Xtraininfo_new
        Ytrain = Ytrain_new
        Xremain = Xremain_new.reset_index(drop=True)
        Xremaininfo = Xremaininfo_new.reset_index(drop=True)
        Yremain = Yremain_new.reset_index(drop=True) ##
        
        Xinfoexplored.append(xinf_next_best)
        Yexplored.append(y_next_best)
        
    XtraininfoInit = [i for i in XtraininfoInit.values]
    YtrainInit = [i for i in YtrainInit.values]
    Xinfoexplored = [i for i in Xinfoexplored]
    Yexplored = [i for i in Yexplored] 
    
    # concatenation combines elements of the lists without the extra tags like 'dtype' and 'Potential' 
    result = np.concatenate((YtrainInit,Yexplored,XtraininfoInit,Xinfoexplored))
    print(result)
        
    return xoptval_int, yoptval_int, xoptval, yoptval, mae, rmse, result

file_name = "Bayesian_Optimization" 
file = open(file_name+'.txt', 'a+')
file.write(f'Shape of original X and Y data:')
file.write(f'{Xdata.shape}, {Ydata.shape}')
file.close() 

n_train = 5 
n_iterations = 10
n_runs = 2

maes = np.empty(shape=(n_runs, n_iterations))
rmses = np.empty(shape=(n_runs, n_iterations))
results = pd.DataFrame(dtype=object)
#results = np.empty(shape=(n_runs, 2 * (n_iterations + n_train))) ## not sure if the second index should be 1 or 2 * (n_iterations + n_train) 
for i in range(n_runs): 
    xoptval_int, yoptval_int, xoptval, yoptval, mae, rmse, result = find_best_mol(Xdata, Xdatainfo, Ydata, n_iterations=n_iterations)
    
    # gets rid of the extra bracket around each item that (potentially) prevented it from going through logic conditions 
    result = np.squeeze(result)
    
    file = open(file_name+'.txt', 'a+')
    file.write(f'The best HBE {xoptval} has potential {yoptval}')
    np.concatenate((maes, [mae]))
    np.concatenate((rmses, [rmse]))
    #np.concatenate((results, [result]))
    #np.append(results, result)
    results = results.append(pd.DataFrame(result))
                              
    # gets rid of the extra bracket around each item that (potentially) prevented it from going through logic conditions 
    result = np.squeeze(result)
    
    print(f"Run {i}, HBEs with potential of 1.55 +- 0.15: ") 
    
    for ii, x in enumerate(result):
        if isinstance(x, float):
            if np.abs(x) < 0.15: 
                ## the enumeration of results array seems weird, the below returns a random reduction potential instead of corresponding SMILES
                print(result[ii + n_train + n_iterations]) 

pd.DataFrame(maes).to_csv('BO_Mean_Absolute_Error.csv')
pd.DataFrame(rmses).to_csv('BO_Root_Mean_Squared_Error.csv')
pd.DataFrame(results).to_csv('BO_Results.csv', quoting=csv.QUOTE_NONNUMERIC)
```
